### Завдання 6 

Напишіть функції, що:
* рохраховує мінімальне та максимальне значення
* середнє значення послідовності
* медіану послідовності
* частоту кожної цифри в послідовності
* середнє квадратичне відхилення

Послідовність чисел ганеруємо рандомно

```python
import random
import math

#генеруэмо тисячу чисел в діапазоні від 0 до 100
list_with_random_numbers = [random.randint(0,100) for i in range(1000)]

list_with_random_numbers                   
```

```python
#Функція, що рохраховує мінімальне значення

def min_value(x):
    min_value= x[0] 
    
    for i in x:
        if i < min_value:
            min_value= i
            
    return min_value

min_numbers= min_value(list_with_random_numbers)

print (f'min_numbers: {min_numbers}')
```

```python
#альтернатива (мінімальне значення)

min_numbers= min(list_with_random_numbers)

print (f'min_numbers: {min_numbers}')
```

```python
#Функція, що рохраховує максимальне значення:

def max_value(x):
    max_value= x[0]
    
    for i in x:
        if i > max_value:
            max_value= i
            
    return max_value

max_numbers= max_value(list_with_random_numbers)

print (f'max_numbers: {max_numbers}')
```

```python
#альтернатива (максимальне значення)

max_numbers= max(list_with_random_numbers)

print (f'max_numbers: {max_numbers}')
```

```python
#Функція, що рохраховує середнє значення послідовності

def avg_value(x):
    s= 0
    
    for i in x:
        s=s+i
    avg_value= s/len(x)
    
    return avg_value

avg_numbers= avg_value(list_with_random_numbers)

print (f'avg_numbers: {avg_numbers}')
```

```python
#альтернатива (середнє значення)

import numpy            #можно import statistics

avg_numbers= numpy.mean(list_with_random_numbers)

print (f'avg_numbers: {avg_numbers}')
```

```python
#Функція, що рохраховує медіану послідовності

def median_value(x):   
    x.sort()
    
    if len(x) % 2 == 0:
        median_value = (x[len(x)//2 - 1] + x[len(x)//2]) / 2
    else:
        median_value = x[len(x)//2]
    return median_value    

median_numbers= median_value(list_with_random_numbers)

print (f'median_value: {median_numbers}')
```

```python
#альтернатива (медіана)

import numpy            #можно import statistics

median_numbers= numpy.median(list_with_random_numbers)

print (f'median_value: {median_numbers}')
```

```python
#Функція, що рохраховує частоту кожної цифри в послідовності

def often_velue(x):         #створюємо функцію, що буде додавати значення у словник
    
    dict_velue = {}      #cтворюємо словник, де ключем буде x, значенням - частота зустрічання x
    
    for i in x:
        if i in dict_velue:
            dict_velue[i] = dict_velue[i] + 1 #якщо цифра вже є у словнику то додаємо 1
        
        else:
            dict_velue[i] = 1 #якщо цифра попадає в словник вперше - присвоюємо 1
            
    return dict_velue
        
often_numbers= often_velue(list_with_random_numbers)

print (f'often_numbers: {often_numbers}')
```

```python
#Функція, що рохраховує середнє квадратичне відхилення (середню величину, на яку відрізняються дані від середнього арифметичного)

def standard_deviation_velue(x):

    s = 0
    for i in x:
        s = s + (i - (sum(x)/len(x)))**2 #рахуємо суму піднесених до квадрату різниць (від кожного елементу віднімаємо середнє значення з усіх елементів))
        
        dispersion_= s/(len(x) - 1) #вираховуємо дисперсію для вибірки
        
        standard_deviation_velue= dispersion_**(1/2)
        
    return standard_deviation_velue

standard_deviation_numbers= standard_deviation_velue(list_with_random_numbers)

print (f'standard_deviation_numbers: {standard_deviation_numbers}')
```

```python
#альтернатива (середнє квадратичне відхилення)

import statistics

standard_deviation_numbers= statistics.stdev(list_with_random_numbers)

print (f'standard_deviation_numbers: {standard_deviation_numbers}')
```
